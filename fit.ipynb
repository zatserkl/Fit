```python
# https://micropore.wordpress.com/2017/02/04/python-fit-with-error-on-y-axis/

--> DO NOT USE THIS CELL !!!

import numpy as np
from pylab import *
from scipy.optimize import curve_fit

def func(x, a, b, c):
    return a * x *x + b*x + c

# test data and error
x = np.linspace(-10, 10, 100)
y0 = - 0.07 * x * x + 0.5 * x + 2.
noise = np.random.normal(0.0, 1.0, len(x))
y = y0 + noise

# curve fit [with only y-error]
popt, pcov = curve_fit(func, x, y, sigma=1./(noise*noise))
perr = np.sqrt(np.diag(pcov))

#print fit parameters and 1-sigma estimates
print("fit parameter 1-sigma error")
print("———————————–")
for i in range(len(popt)):
    print(str(popt[i])+" +- "+str(perr[i]))

# prepare confidence level curves
nstd = 5. # to draw 5-sigma intervals
popt_up = popt + nstd * perr
popt_dw = popt - nstd * perr

fit = func(x, *popt)
fit_up = func(x, *popt_up)
fit_dw = func(x, *popt_dw)

#plot
fig, ax = plt.subplots(1)
rcParams["xtick.labelsize"] = 18
rcParams["ytick.labelsize"] = 18
rcParams["font.size"] = 20
errorbar(x, y0, yerr=noise, xerr=0, hold=True, ecolor="k", fmt="none", label="data")

xlabel("x", fontsize=18)
ylabel("y", fontsize=18)
title("fit with only Y-error", fontsize=18)
plot(x, fit, "r", lw=2, label="best fit curve")
plot(x, y0, "k-", lw=2, label="True curve")
ax.fill_between(x, fit_up, fit_dw, alpha=.25, label="5-sigma interval")
legend(loc="lower right",fontsize=18)
```

```python
#
#   Fit Gauss
#   No errors, no initialization
#

import numpy as np
import matplotlib.pyplot as plt
from scipy.optimize import curve_fit

def Gauss(x, A, mean, sigma):
    return A * np.exp(-(x - mean)**2 / 2 / sigma**2)

np.random.seed(1)

mu, sigma = 0, 1
rnormal = np.random.normal(mu, sigma, 1000)

fig, ax = plt.subplots()

bins = np.arange(-5*sigma, 5*sigma, 0.2)        # set step 0.2
# bins = np.linspace(-5*sigma, 5*sigma, 100+1)  # set the number of points (100)

y, x, _ = ax.hist(rnormal, bins=40, histtype="step")
x = 0.5 * (x[1:] + x[:-1])

popt, pcov = curve_fit(Gauss, x, y)
popt[2] = abs(popt[2])  # positive sigma
print(f"popt: {popt}")

ax.plot(x, Gauss(x, *popt))
ax.grid()

text = f"A = {popt[0]:.0f}\n$\mu$ = {popt[1]:.2f}\n$\sigma$ = {popt[2]:.2f}"
ax.text(0.05, 0.95, text, transform=ax.transAxes, fontsize=14, va='top')
```

```python
#
#   Fit Gauss
#   No errors, no initialization
#

import numpy as np
import matplotlib.pyplot as plt
from scipy.optimize import curve_fit

def fGauss(x, A, mean, sigma):
    return A * np.exp(-(x - mean)**2 / 2 / sigma**2)

np.random.seed(1)

mu, sigma = 0, 1
rnormal = np.random.normal(mu, sigma, 1000)

plt.figure()

bins = np.arange(-5*sigma, 5*sigma, 0.2)        # set step 0.2
# bins = np.linspace(-5*sigma, 5*sigma, 100+1)  # set the number of points (100)

yfit, xfit, _ = plt.hist(rnormal, bins=40, histtype="step")
xfit = 0.5 * (xfit[1:] + xfit[:-1])

popt, pcov = curve_fit(fGauss, xfit, yfit)
popt[2] = abs(popt[2])  # positive sigma
print(f"popt: {popt}")

chi2ndf = np.sum((yfit - fGauss(xfit, *popt))**2) / (len(xfit) - 3)
err = np.sqrt(yfit)
err[err == 0] = 1
chi2ndf = np.sum(((yfit - fGauss(xfit, *popt)) / err)**2) / (len(xfit) - 3)
print(f'chi2ndf: {chi2ndf}')
plt.plot(xfit, fGauss(xfit, *popt))
plt.grid()

text = f"A = {popt[0]:.0f}\n$\mu$ = {popt[1]:.2g}\n$\sigma$ = {popt[2]:.2g}\n$\chi^2$ = {chi2ndf:.2f}"
plt.text(0.05, 0.95, text, transform=plt.gca().transAxes, fontsize=14, va='top')
```

```python
#
#   Fit Gauss
#   No errors
#

import numpy as np
import matplotlib.pyplot as plt
from scipy.optimize import curve_fit

class FitGauss:
    @staticmethod
    def fGauss(x, A, mean, sigma):
        return A * np.exp(-(x - mean)**2 / 2 / sigma**2)
    
    def __init__(self, xfit, yfit):
        self.xfit = xfit
        self.yfit = yfit
    
    def fit(self):
        # estimate parameters of Probability Density Function assumming that it's Gaussian

        self.total_events = np.sum(self.yfit)  # total number of events

        self.mean0 = np.sum(self.xfit * self.yfit) / self.total_events
        var = np.sum((self.xfit - self.mean0)**2 * self.yfit) / self.total_events
        self.sigma0 = np.sqrt(var)

        w = xfit[1] - xfit[0]  # area occupaed by 1 event is w*1. Total area is total_events*w square units.
        self.A0 = self.total_events * w / np.sqrt(2 * np.pi * var)  # events per bin in maximum

        self.p0 = [self.A0, self.mean0, self.sigma0]

        self.popt, self.pcov = curve_fit(FitGauss.fGauss, self.xfit, self.yfit, p0=self.p0)
        self.popt[2] = abs(self.popt[2])  # positive sigma

        self.chi2ndf = np.sum((yfit - FitGauss.fGauss(xfit, *self.popt))**2) / (len(xfit) - 3)
        err = np.sqrt(yfit)
        err[err == 0] = 1
        self.chi2ndf = np.sum(((yfit - FitGauss.fGauss(xfit, *self.popt)) / err)**2) / (len(xfit) - 3)

        return self.popt, self.pcov

np.random.seed(1)

mu, sigma = 0, 1
rnormal = np.random.normal(mu, sigma, 10000)

plt.figure()

bins = np.arange(-5*sigma, 5*sigma, 0.2)        # set step 0.2
# bins = np.linspace(-5*sigma, 5*sigma, 100+1)  # set the number of points (100)

yfit, xfit, _ = plt.hist(rnormal, bins=40, histtype="step")
xfit = 0.5 * (xfit[1:] + xfit[:-1])

# i1, i2 = 970, 990         # x-index range of camera pixels
# xfit = np.arange(i1, i2)
# yfit = prof_y[i1:i2]
# print(f'len(xfit): {len(xfit)}, len(yfit): {len(yfit)}')

# Estimate mean and sigma from the histogram - probability density function

total_events = np.sum(yfit)
print(f'total_events: {total_events}')

est_mean = np.sum(xfit * yfit) / total_events
print(f'est_mean: {est_mean}')

est_var = np.sum((xfit - est_mean)**2 * yfit) / total_events
est_sigma = np.sqrt(est_var)
print(f'est_sigma: {est_sigma}')

w = xfit[1] - xfit[0]  # area occupaed by 1 event is w*1
est_A = total_events * w / np.sqrt(2 * np.pi * est_var)  # events per bin in maximum
print(f'est_A: {est_A}')

gauss = FitGauss(xfit, yfit)
gauss.fit()
popt, pcov, chi2ndf = gauss.popt, gauss.pcov, gauss.chi2ndf

print(f'p0: {gauss.p0}')
print(f'popt: {popt}')
print(f'pcov: {pcov}')
print(f'chi2ndf: {chi2ndf}')

plt.plot(xfit, FitGauss.fGauss(xfit, *popt))
plt.grid()

text = f"A = {popt[0]:.0f}\n$\mu$ = {popt[1]:.3f}\n$\sigma$ = {popt[2]:.3f}\n$\chi^2$ = {chi2ndf:.2g}"
plt.text(0.05, 0.95, text, transform=plt.gca().transAxes, fontsize=14, va='top')
```
